### An example of the scraped image

```python

import tarfile
import json
from PIL import Image

# Path to the tar file
tar_file_path = os.path.join("artwork_picture", '00000.tar')

# Open the tar file
with tarfile.open(tar_file_path, 'r') as tar:
    # List all files in the tar archive
    tar_members = tar.getmembers()
    
    # Extract and display the first three files
    for member in tar_members[:3]:
        if member.isfile():
            file = tar.extractfile(member)
            content = file.read()
            if member.name.endswith('.jpg'):
                # Display the image
                image = Image.open(file)
                display(image)
            elif member.name.endswith('.json'):
                # Display the JSON content
                json_content = json.loads(content)
                print(json.dumps(json_content, indent=4))
            elif member.name.endswith('.txt'):
                # Display the text content
                print(content.decode('utf-8'))
```
