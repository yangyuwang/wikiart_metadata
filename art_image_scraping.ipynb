```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}
```

```python
import requests
from tqdm import tqdm
import urllib.request

url_list = artwork_data["image_url"].tolist()

if "image_n" not in artwork_data.columns:
    artwork_data["image_n"] = None


found_list= os.listdir("/Volumes/SSD/存储/artwork_picture")

for idx, url in tqdm(enumerate(url_list), total=len(url_list), unit="artwork"):
    try:
        if str(idx) + ".jpg" not in found_list:
            req = urllib.request.Request(url, headers=headers)
            response = urllib.request.urlopen(req, timeout=10)
            if response.getcode() >= 400:
                raise urllib.error.HTTPError(
                    response.url, response.getcode(), response.msg, response.headers, None
                )

            with open(f"/Volumes/SSD/存储/artwork_picture/{idx}.jpg", "wb") as file:
                file.write(response.read())

            artwork_data.loc[artwork_data["image_url"] == url, "image_n"] = idx

    except urllib.error.HTTPError as e:
        print(f"下载失败: {url} - HTTP错误: {e}")

    except urllib.error.URLError as e:
        print(f"下载失败: {url} - URL错误: {e}")

    except Exception as e:
        print(f"未知错误: {url} - 错误: {e}")
```

```python
artwork_data.to_csv("artwork_data/artwork_data_image_done.csv", index=False)
```
